#### Import Libraries

```python
import os
import requests
import pandas as pd
import random
import shutil
from tqdm import tqdm
```

#### Download ISIC Metadata

```python
metadata_url = "https://challenge2020.isic-archive.com/api/v1/dataset/isic_2020/metadata"

metadata = pd.read_json(metadata_url)

metadata.head()
```

#### Choose a Random Sample

```python
sample_size = 2000
sample_metadata = metadata.sample(sample_size, random_state=42)

sample_metadata.head()
```

#### Download Only the Sample Images

```python
os.makedirs("sample_images", exist_ok=True)

for img_id in tqdm(sample_metadata["image"]):
    
    url = f"https://isic-archive.com/api/v1/image/{img_id}/download"
    
    path = f"sample_images/{img_id}.jpg"
    
    if not os.path.exists(path):
        r = requests.get(url)
        with open(path, "wb") as f:
            f.write(r.content)
```

#### Save Metadata for the Sample

```python
sample_metadata.to_csv("sample_metadata.csv", index=False)
```

#### Zip the Sample for Sharing

```python
shutil.make_archive("isic_sample_dataset", "zip", "sample_images")
```
